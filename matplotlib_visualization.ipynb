# Plotting x and y points
The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

If we need to plot a line from (1, 3) to (8, 10), we have to pass two arrays [1, 8] and [3, 10] to the plot function.

```python
x_points = np.array([3, 80])
y_points = np.array([10, 100])
plt.plot(x_points, y_points, 'o')
plt.show()
```

# Create more `plot` using `for` loop operations

```python
import pandas as pd 

data = np.random.rand(16, 12)
df = pd.DataFrame(data)

# Create a 4x4 subplot grid
fig, axes = plt.subplots(4, 4, figsize=(10, 10))

# Generate a list of colors for each subplot
colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray',
          'cyan', 'lime', 'magenta', 'yellow']

# Iterate through each subplot and plot data with different color
for i, ax in enumerate(axes.flatten()):
    if i < df.shape[1]:
        ax.plot(df.iloc[:, i], color=colors[i])
        ax.set_title(f'Plot {i+1}', fontsize=8)
    else:
        ax.axis('off')  # Turn off extra subplots if there are fewer columns than subplots

# Adjust the spacing between subplots
fig.tight_layout()

# Show the plot
plt.show()
```
