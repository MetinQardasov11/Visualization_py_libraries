# `Load data` 
# and `Find` 
`info(), describe(), describe() object` 
# using class

```python
class data_information():
    
    def __init__(self,file_name):
        self.file_name = file_name
        self.data = None
        
    def load_data(self):
        try:
            self.data = pd.read_csv('country.csv')
            print(f'{self.file_name} successfully loaded!!!')
        except FileNotFoundError:
            print(f'{self.file_name} not found!!!')
    
    def info_data(self):
        if self.data is not None:
            print('Data statistics...')
            print(self.data.info())
        else:
            print('Data is empty!!!')
    
    def describe_data(self):
        if self.data is not None:
            print('Data statistics...')
            print(self.data.describe())
        else:
            print('Data is empty!!!')
            
    def describe_object_data(self):
        if self.data is not None:
            print('Data statistics...')
            print(self.data.describe(include='O'))
        else:
            print('Data is empty!!!')
```

# `Input` data

```python
if __name__ == "__main__":
    data = data_information('country.csv')
    data.load_data()
    data.info_data()
    data.describe_data()
    data.describe_object_data()
```
